```python
import random
import matplotlib.pyplot as plt
```

# Sierpinski Carpet

The Sierpinski Carpet or Sierpinski Triangle can be computer generated by a fairly simple algorithm. The idea behind it is:

pick three points (defining the triangle you want to generate)
from an arbitrary starting point, pick a vertex at random and move half the distance to that vertex and repeat starting from this newly defined point. Below, this algorithm is implemented for 50,000 iterations, starting at a point to the upper right of the three vertices.

```python
pnt1 = (0, 0)
pnt2 = (250, 100)
pnt3 = (100, 250)
sierpinski_carpet = [pnt1, pnt2, pnt3]

old_point = (250, 250)
for _ in range(50000):
    rand_int = random.randint(1, 3)
    if rand_int == 1:
        target = pnt1
    elif rand_int == 2:
        target = pnt2
    else:
        target = pnt3
    new_point = (
        round(old_point[0] - (old_point[0] - target[0]) / 2),
        round(old_point[1] - (old_point[1] - target[1]) / 2),
    )
    old_point = new_point
    sierpinski_carpet.append(new_point)
```

```python
xs, ys = zip(*sierpinski_carpet)
plt.figure(figsize=(6,6))
plt.scatter(xs, ys, s=0.05, color='black')
plt.axis('equal')
plt.show()
```
